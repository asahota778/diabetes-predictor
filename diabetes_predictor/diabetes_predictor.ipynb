```python


class Preprocessor_fill:
    def __init__(self):
        self.columns_to_fill = ['height', 'weight']  # Columns to fill with mean

    def fill_nan_with_mean(self, df):
        # For each column in the columns_to_fill, replace NaN values with the mean
        for column in self.columns_to_fill:
            print(f"Processing column: {column}")

            # Print the column data before processing (for debugging purposes)
            print(f"Before conversion to numeric: \n{df[column].head()}")

            # Convert the column to numeric (coerce any non-numeric values to NaN)
            df[column] = pd.to_numeric(df[column], errors='coerce')

            # Print the column data after conversion to numeric (for debugging purposes)
            print(f"After conversion to numeric: \n{df[column].head()}")

            # Calculate the mean of the column
            mean_value = df[column].mean()
            print(f"Mean of '{column}': {mean_value}")  # Debugging print

            # Fill NaN values with the column mean
            df[column] = df[column].fillna(mean_value)

            # Print the column data after filling NaN values
            print(f"After filling NaN: \n{df[column].head()}")
        
        return df

preprocessors_fill = Preprocessor_fill()
```

```python
train_df_clean = preprocessors_fill.fill_nan_with_mean(train_df)
```

```python
# Assuming train_df_clean is already defined

# Count the number of NaN values in each column
null_count = train_df_clean.isna().sum()

# Get the total number of rows in the DataFrame
total_rows = train_df_clean.shape[0]

# Display the null count and total number of rows
print("Null count for each column:")
print(null_count)

print(f"\nTotal number of rows in the DataFrame: {total_rows}")
```
